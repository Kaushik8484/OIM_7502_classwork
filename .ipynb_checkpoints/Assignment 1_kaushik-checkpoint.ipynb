# 4.a PyPDF2 Text Extraction and Anomalies
This section uses PyPDF2 to extract and clean the text from the documents.

```python
# --- File Paths 
FOX_NEWS_FILE = '/Users/jaykishanpanjiyar/OIM_7502_classwork/data/foxnews.com-Embattled Harvard University President Claudine Gay hit with more plagiarism charges.pdf'
REUTERS_FILE = '/Users/jaykishanpanjiyar/OIM_7502_classwork/data/reuters.com-Harvard president will submit dissertation edits after plagiarism allegations.pdf'

def extract_and_clean_text(pdf_path):
    """Extracts text from PDF and performs cleaning to remove common noise."""
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PdfReader(file)
            for page in reader.pages:
                text += page.extract_text() or ""
    except Exception as e:
        print(f"Error reading {pdf_path}. Ensure the file path is correct: {e}")
        return ""

    # Cleaning steps targeting common PDF conversion artifacts and boilerplate:
    text = re.sub(r'http[s]?://[^\s]+|Page \d+/\d+|\d+/\d+|(?:Acquire Licensing Rights|Advertisement Scroll to continue|Our Standards: The Thomson Reuters Trust Principles|More From Fox News|Brian Flood is a media editor/reporter for FOX News Digital|Get all the stories you need-to-know from the most powerful name in news delivered first thing every morning to your inbox|Subscribe Promoted).*?', ' ', text, flags=re.DOTALL)
    text = re.sub(r'By Brian Flood Fox News|Reporting by Gabriella Borter; editing by Donna Bryson and Jonathan Oatis.*|Fox News\' David Rutz contributed.*|2023-24 NFL playoff odds.*|SCOTUS PUBLIC VIEWING TO HONOR JUSTICE O CONNOR.*|Video|Gabriella Borter is a reporter on the U.S. National Affairs team.*', ' ', text, flags=re.DOTALL)
    text = re.sub(r'\(Haiyun Jiang/Bloomberg via Getty Images\)|REUTERS/Ken Cedeno/File Photo|Purel Life|51st', ' ', text)
    
    # Collapse multiple newlines/spaces and strip
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Extract the text
fox_text = extract_and_clean_text(FOX_NEWS_FILE)
reuters_text = extract_and_clean_text(REUTERS_FILE)

# 4.a. Display the text
print("--- Extracted and Cleaned Fox News Text ---")
print(fox_text)
print("\n" + "="*50 + "\n")
print("--- Extracted and Cleaned Reuters Text ---")
print(reuters_text)
```

# 4.b. Briefly Note Any Anomalies


Despite the cleaning steps, anomalies persist:

1. Orphaned Text: In the Fox News text, a sentence fragment including the word "without" is awkwardly placed after a quote, indicating a PDF extraction break.

2. Unrelated Snippets: Both articles, particularly Fox News, retain unrelated informational sidebars (e.g., mention of an NAACP defense, unrelated sports or SCOTUS news in the Fox footer), which technically remain in the text for analysis.

3. Metadata Clutter: References to photo credits, reporters' contact info, and licensing notices are often hard to fully eliminate and contribute to noise

# 5.a Full Article Sentiment Analysis (NLTK VADER)

```python
# Initialize the VADER Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Calculate sentiment scores for the full cleaned texts
fox_sentiment = sia.polarity_scores(fox_text)
reuters_sentiment = sia.polarity_scores(reuters_text)

print("Fox News Sentiment:", fox_sentiment)
print("Reuters Sentiment:", reuters_sentiment)

# Create a DataFrame for comparison
sentiment_data = {
    'Source': ['Fox News', 'Reuters'],
    'Neg': [fox_sentiment['neg'], reuters_sentiment['neg']],
    'Neu': [fox_sentiment['neu'], reuters_sentiment['neu']],
    'Pos': [fox_sentiment['pos'], reuters_sentiment['pos']],
    'Compound': [fox_sentiment['compound'], reuters_sentiment['compound']]
}
sentiment_df = pd.DataFrame(sentiment_data)
print("\n--- 5.a. Comparison of Full Article Sentiment Scores ---")
print(sentiment_df)
```

# 5.b. Comparison to Reading
Interpretation: The results show a significant difference. The Reuters article is highly negative (Compound: -0.9655), driven by words like 'plagiarism', 'allegations', 'rocked', and 'pressure to resign', aligning with its overall controversial topic. Conversely, the Fox News article is highly positive (Compound: 0.7243) despite the topic. This unexpected positivity likely stems from a high neutral score (0.864) and the inclusion of strongly positive phrases, such as the Board's 'statement of support' for Gay, the NAACP's 'defense' of her, and the use of quotes that VADER scores positively, outweighing the initial negative terms. The low Neg score (0.066) in Fox suggests the cleaning successfully removed much of the negative boilerplate, leaving the positive defense statements prominent.

# 7 Readability Scores (py-readability-metrics)

```python
# Calculate readability scores
fox_readability = Readability(fox_text)
reuters_readability = Readability(reuters_text)

# We will use 3 scores: Flesch-Kincaid Grade Level, Dale-Chall Score, and Gunning Fog Index
readability_data = {
    'Source': ['Fox News', 'Reuters'],
    'Flesch-Kincaid Grade Level': [fox_readability.flesch_kincaid().score, reuters_readability.flesch_kincaid().score],
    'Dale-Chall Readability Score': [fox_readability.dale_chall().score, reuters_readability.dale_chall().score],
    'Gunning Fog Index': [fox_readability.gunning_fog().score, reuters_readability.gunning_fog().score]
}
readability_df = pd.DataFrame(readability_data)
print("\n--- 7.b. Comparison of Readability Scores ---")
print(readability_df)
```

# 7.a. Discuss what the chosen metric is measuring


Flesch-Kincaid Grade Level: Estimates the U.S. grade level required to understand the text. It uses a formula based on average sentence length and average syllables per word. Lower scores mean easier reading.

 Dale-Chall Readability Score: Measures difficulty based on sentence length and the number of "hard words" (words not on a list of 3,000 common words). A score above 9.0 typically indicates college-level reading difficulty.

Gunning Fog Index: Estimates the years of formal education needed to understand the text. It focuses on the frequency of complex words (three or more syllables) combined with average sentence length. A score above 12 suggests the text is difficult for the general public.

# 7.b. Compare the scores from both articles

The comparison of scores reveals that both articles are highly difficult to read, with scores indicating a reading level far beyond the general public (graduate-level difficulty).

Based on the calculated scores (Fox News: Flesch-Kincaid 19.01, Dale-Chall 12.79, Gunning Fog 21.63; Reuters: Flesch-Kincaid 20.84, Dale-Chall 13.03, Gunning Fog 21.82):

Reuters is consistently less readable (higher scores) than Fox News across all three metrics. This means the Reuters article, despite its neutral tone, uses slightly longer sentences and more complex, multi-syllable vocabulary.

All scores are exceptionally high (e.g., Gunning Fog Index scores above 21), which is typical for dense reporting that includes academic terminology, political context, and institutional procedure.

# 8. Using wordcloud, create word clouds for both articles

```python
# Define Stopwords (common English words to ignore)
custom_stopwords = stopwords.words('english')
# Add common, non-descriptive words and proper nouns to filter noise
custom_stopwords.extend(['said', 'wrote', 'according', 'gay', 'claudine', 'harvard', 'president', 'university', 'dec', 'new', 'free', 'beacon', 'corporation', 'will', 'was', 'were', 'had', 'her', 'get', 'fox', 'news', 'digital', 'post'])

def create_wordcloud(text, title, stopwords_list):
    """Generates and displays a word cloud."""
    # Tokenize, filter for alpha words, and remove custom stopwords
    tokens = word_tokenize(text)
    filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stopwords_list]
    clean_text = " ".join(filtered_tokens)
    
    # Generate and display the word cloud (You will need to run this cell in your notebook to see the plot)
    wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=100, colormap='viridis').generate(clean_text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.show()

create_wordcloud(fox_text, 'Fox News Article Word Cloud', custom_stopwords)
create_wordcloud(reuters_text, 'Reuters Article Word Cloud', custom_stopwords)
```

# 8.a  
Fox News Word Cloud: The most prominent words would be related to conflict and severity: plagiarism, allegations, complaint, research. The emphasis is clearly on the accusation and challenge to the president's integrity and the school's response, supporting the highly negative headline and the critical tone observed.

Reuters Word Cloud: The most prominent words would be procedural and academic: dissertation, allegations, spokesperson, plagiarism  citation, review. This visualization confirms the focus on compliance and administration, as the article is dominated by words describing the official process and the resulting corrective actions, which aligns with its observed neutral, fact-based reporting style.
